```python
import os
import pandas as pd
from bertopic import BERTopic

# Define the path to the folder containing the CSV files
data_folder = r"C:\Users\ekizi\Downloads\haberler"

# Initialize an empty list to store the articles
articles = []

# Loop through the CSV files in the folder
for filename in os.listdir(data_folder):
    if filename.endswith(".csv"):
        file_path = os.path.join(data_folder, filename)
        data = pd.read_csv(file_path)
        articles.extend(data["Sample"].tolist())

# Create BERTopic model
model = BERTopic()

# Fit the model to the data
topics, _ = model.fit_transform(articles)

# Get the most frequent topics
top_topics = model.get_topic_freq()

# Print the top topics
for topic in top_topics:
    print(f"Topic {topic[0]}: {topic[1]} documents")
```
